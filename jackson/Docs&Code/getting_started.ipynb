```python
import pandas as pd
import altair as alt
from statsmodels.tsa.vector_ar.var_model import VAR
from random import random
```

```python
cars = pd.read_csv("../Data/auto-mpg.csv")
```

```python
cars.sample(5)
```

```python
cars.describe()
```

```python
cars[pd.isnull(cars["horsepower"])]
```

```python
alt.Chart(cars).mark_bar().encode(
    x = alt.X("mpg",bin=True),
    y = "count()"
)
```

```python
alt.Chart(cars).mark_point().encode(
    x = "horsepower",
    y = "mpg"
)
```

```python
data = list()
for i in range(100):
    v1 = i + random()
    v2 = v1 + random()
    row = [v1, v2]
    data.append(row)
```

```python
# fit model
model = VAR(data)
model_fit = model.fit()
# make prediction
yhat = model_fit.forecast(model_fit.y, steps=1)
print(yhat)
```
